```python
import requests
import json
from pandas.io.json import json_normalize
#Reading data from json directly and setting max limit to 10000 (Note there are only 100 values )
url = "https://www.playerauctions.com/api/csgo/ranking?pageindex=1&filter=top&sort=win&limit=10000"

response = requests.get(url)
data = json.loads(response.text)
df1=json_normalize(data[1])
df1
```

```python
#This is data of top 3 players
df2=json_normalize(data[0])
df2
```

```python
df = pd.concat([df2, df1], axis=0)
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={'englishName': 'Country'}, inplace=True)
df.drop(columns=['avatar','windowsBroken','weaponsToTeam'],inplace=True)
```

```python
df.to_csv('CSGO_2023_players.csv')
```
