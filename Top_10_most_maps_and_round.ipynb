```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('CSGO Player Dataset.csv')
```

```python
df.head()
```

```python
most_maps = df.sort_values(by=['Number of Maps Played'],ascending=False)
```

```python
most_maps = most_maps.iloc[0:10,:]
```

```python
most_rounds = df.sort_values(by=['Number of Rounds Played'],ascending=False)
```

```python
most_rounds = most_rounds.iloc[0:10,:]
most_rounds
```

```python
most_maps
```

```python
corr = most_maps['Number of Maps Played'].corr(most_rounds['Number of Rounds Played'])
```

```python
corr
```

## Corrrelation Between No. of Maps Played and No. of Rounds Played

- As we can see that both attributes are strongly correlated to each other, which mean as the player who played more maps must have also played more number of rounds as well
- KD ratio is also correlated to HLTV Ratings

```python
# most_maps[['Number of Maps Played','Number of Rounds Played']].corr()
sns.heatmap(most_maps[['Number of Maps Played','Number of Rounds Played']].corr(),annot=True,cmap="Blues")
```
