```python
import pandas as pd 
import numpy as np
df = pd.read_csv('CSGO Player Dataset.csv')
def deaths(kdd,kdr):
    #Kill death difference
    #Kills/Death Ratio
    if(kdr!= 1):
        d = int(kdd/(kdr-1))
    else:
        d = np.nan
    return d
df['Total Deaths'] = df[['Kill Death Difference','K/D Ratio']].apply(lambda df:deaths(df['Kill Death Difference'],df['K/D Ratio']),axis=1)
df['Total Kills'] = df['Kill Death Difference'] + df['Total Deaths']
df.head(5)
```
