```python
import pandas as pd
import plotly.offline as po
import plotly.graph_objs as pgo
df=pd.read_csv('HLTV.csv')
data=[]
for i in df['Name']:
    trace=pgo.Scatter(x=df['Name'],y=df['HLTV Rating'],mode='lines',name=i)
data.append(trace)
layout=pgo.Layout(title='Top 10 players based on HLTV Rating',xaxis=dict(title='Names'),yaxis=dict(title='HLTV Rating'))
fig=pgo.Figure(data=data,layout=layout)

po.plot(fig)
```

```python
import pandas as pd
import plotly.offline as po
import plotly.graph_objs as pgo
df=pd.read_csv('KD Ratio.csv')
data=[]
for j in df['Name']:
    trace=pgo.Scatter(x=df['Name'],y=df['K/D Ratio'],mode='lines',name=j)
data.append(trace)
layout=pgo.Layout(title='KD Ratio variation',xaxis=dict(title='TOP 10 PLAYERS BASED ON K/D RATIO'),yaxis=dict(title='KD Ratio Rating'))
fig2=pgo.Figure(data=data,layout=layout)

po.plot(fig2)
```
